## Handling outliers

```python
df.boxplot(column=['Generosity'])
```

```python
def removeOutliers(col):
    sorted(col)
    q1,q2 = col.quantile([0.25,0.75])
    IQR = q2 - q1
    lower = q1 - (1.5 * IQR)
    upper = q2 + (1.5 * IQR)
    return lower, upper
```

```python
lowScore, highScore = removeOutliers(df['Generosity'])
new_df2 = df.copy()
new_df2['Generosity'] = np.where(new_df2['Generosity']>highScore, highScore, new_df2['Generosity'])
new_df2['Generosity'] = np.where(new_df2['Generosity']<lowScore, lowScore, new_df2['Generosity'])
```

```python
new_df2.boxplot(column=['Generosity'])
```
